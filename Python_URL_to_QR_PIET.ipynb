```python
# taking url as input
url = input("Enter your  URL Here")
# taking image which you wants
# in the QR code center
logo = Image.open(img_url)
basewidth = 100
# adjust image size
wpercent = (basewidth/float(logo.size[0]))
hsize = int((float(logo.size[1])*float(wpercent)))
logo = logo.resize((basewidth, hsize), Image.ANTIALIAS)
QRcode = qrcode.QRCode(error_correction=qrcode.constants.ERROR_CORRECT_H)
QRcode.add_data(url)
QRcode.make()
# Assigning color
QRcolor = '#006400'
QRimg = QRcode.make_image(fill_color=QRcolor, back_color="white").convert('RGB')
 
# set size of QR code
pos = ((QRimg.size[0] - logo.size[0]) // 2,
       (QRimg.size[1] - logo.size[1]) // 2)
QRimg.paste(logo, pos)
img = ImageOps.expand(QRimg, border=30, fill=(255,255,255))
draw = ImageDraw.Draw(img)
font = ImageFont.truetype("PTSans.ttf",size=30)
draw.text((80,0),"PIET Summer Internship 2022",(0,0,0),font=font)
img.save('project_QR.png')
display(img)
```
